```python
import pandas as pd, numpy as np, gensim
#!pip install pyldavis
import nltk
import pandas as pd
import numpy as np
from wordcloud import WordCloud
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
amazon = pd.read_csv('C:/Users/Rhishikesh Padole/Downloads/amazon_reviews_big/amazon_reviews_big.csv')
```

```python
docs_clean = []
docs = amazon['reviewText'].fillna('').str.lower().str.replace('[^a-z]', ' ')
stopwords = nltk.corpus.stopwords.words('english')
stopwords.extend(['use', 'good', 'like', '', 'great', 'work', 'one','would', 'mm', 'use', 'take','get', 'work', 'well', 'look'])

stemmer = nltk.stem.PorterStemmer()
```

```python
for doc in docs:
    words = doc.split(' ')
    words_clean = [stemmer.stem(word) for word in words if word not in stopwords]
    docs_clean.append(words_clean)
len(docs_clean)    
```

```python
dictionary = gensim.corpora.Dictionary(docs_clean)
dictionary
```

```python
docs_bow = []
for doc in docs_clean:
    doc_bow = dictionary.doc2bow(doc)
    docs_bow.append(doc_bow)
```

```python
lda_model = gensim.models.LdaModel(docs_bow,
                                  id2word = dictionary,
                                  num_topics = 4)
```

```python
# Checking the total proportion of 1st document :

d = 0
for i in lda_model.get_document_topics(docs_bow[0]):
    d = i[1] + d
print(d)    
```

```python
topics = []

for doc_bow in docs_bow:
    lda_model.get_document_topics(doc_bow)
    doc2topic_prob = pd.DataFrame(doc2topic_prob,
                                 columns = ['topic', 'prob'])
    topic = doc2topic_prob.sort_values('prob', ascending=False)['topic']
    topics.append(topic)
```
